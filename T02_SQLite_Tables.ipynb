```python
str_ = '''SELECT 
            Title,
            name 
        FROM
            albums
        INNER JOIN artists
            USING(ArtistID)
            '''

for row in database.execute(str_):
    print (row)
```

```python
str_ = '''
SELECT
    Name, 
    Title
FROM
    artists
LEFT JOIN albums ON
    albums.ArtistId = artists.ArtistId
ORDER BY Name;
'''

for row in database.execute(str_):
    print (row)
```
